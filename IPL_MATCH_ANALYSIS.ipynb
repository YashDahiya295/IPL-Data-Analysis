```python
#Loading the required libraries 

import pandas as pd
from  matplotlib import pyplot as plt
import seaborn as sns 
```

```python
#Loading the ipl match dataset 
ipl = pd.read_csv('/Users/yashdahiya/Downloads/matches.csv')
```

```python
#having a glance at the first five record of the dataset
ipl.head()
```

```python
#Looking at the numbers of rows and columns in the dataset 
ipl.shape
```

```python
#Getting the frequency of most man of the match awards 
ipl['player_of_match'].value_counts()
```

```python
#getiing top 10 players with the most man of the match 
ipl['player_of_match'].value_counts()[0:10]
```

```python
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
#bar plot always take two parameters one is categorical data and another one is numeric data .
plt.figure(figsize=(8,5)) 
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()) , list(ipl['player_of_match'].value_counts()[0:5]) , color="olive")
plt.show()
```

```python
#Getting the frequency of the result columnn 
ipl['result'].value_counts()
```

```python
#Finding the number of toss wins wrt each team 
ipl['toss_winner'].value_counts()
```

```python
#Extracting the records where a team won batting first 
batting_first = ipl[ipl['win_by_runs']!=0]
```

```python
batting_first
```

```python
#Looking at the head 
batting_first.head()
```

```python
#Making a histogram , For numerical value we have to use histogram 
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after batting first 

batting_first['winner'].value_counts()
```

```python
#Making a bar plot for top 3 teams with most wins after batting first 
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]) ,color=["blue" ,"yellow" ,"red"])
plt.show()
```

```python
#making a pie chart : Here in pie ; the first parameter is categorical data and the second parameter 


plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```

```python
#Making a pie chart for the above scenario 


plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```
